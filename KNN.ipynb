# Reading the dataset


```python
df = pd.read_csv("health.csv")
df.head()
```

```python
df.describe
```

# Splitting the dataset

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train.shape
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import Normalizer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
classifier = KNeighborsClassifier(n_neighbors = 9)
classifier.fit(x_train, y_train)
```

```python
y_test_pred = classifier.predict(x_test)
y_train_pred = classifier.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
test_accuracy = accuracy_score(y_test,y_test_pred)
```

```python
print("Testing Accuracy:",test_accuracy*100)
train_accuracy = accuracy_score(y_train,y_train_pred)
print("Training Accuracy:",train_accuracy*100)
```

# Create a plot to present the performance for different values of k.

```python
acc = []
# Calculating accuracy for K values between 1 and 10
for i in range(1, 11):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train, y_train)
  pred_i = knn.predict(x_test)
  acc.append(accuracy_score(y_test,pred_i))
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), acc, color='red', linestyle='dashed', marker='o',
markerfacecolor='blue', markersize=10)
plt.title('Accuracy vs K-Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
```

# Apply feature scaling- Z-score, Min-Max formula



# Standard Scale

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
clf = KNeighborsClassifier(n_neighbors = 6)
clf.fit(x_train, y_train)
```

```python
y_test_preds = clf.predict(x_test)
y_train_preds = clf.predict(x_train)
tests_accuracy = accuracy_score(y_test,y_test_preds)
print("Testing Accuracy with k=6:",tests_accuracy*100)
trains_accuracy = accuracy_score(y_train,y_train_preds)
print("Training Accuracy with k=6:",trains_accuracy*100)
```

# Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
clf = KNeighborsClassifier(n_neighbors = 6)
clf.fit(x_scaled, y_train)
```

```python
y_test_preds = clf.predict(x_test_scaled)
y_train_preds = clf.predict(x_scaled)
tests_accuracy = accuracy_score(y_test,y_test_preds)
print("Testing Accuracy with k=6:",tests_accuracy*100)
trains_accuracy = accuracy_score(y_train,y_train_preds)
print("Training Accuracy with k=6:",trains_accuracy*100)
```

# Z-score

```python
nscaler = Normalizer()
x_nscaled = nscaler.fit_transform(x_train)
x_test_nscaled = nscaler.fit_transform(x_test)
clf = KNeighborsClassifier(n_neighbors = 6)
clf.fit(x_nscaled, y_train)
```

```python
y_test_preds = clf.predict(x_test_nscaled)
y_train_preds = clf.predict(x_nscaled)
tests_accuracy = accuracy_score(y_test,y_test_preds)
print("Testing Accuracy with k=6:",tests_accuracy*100)
trains_accuracy = accuracy_score(y_train,y_train_preds)
print("Training Accuracy with k=6:",trains_accuracy*100)
```

# Change the values of distance metric and draw comparison graph

```python
acc = []
for i in range(1, 5):
  knn = KNeighborsClassifier(n_neighbors=8,p=i)
  knn.fit(x_train, y_train)
  pred_i = knn.predict(x_test)
  acc.append(accuracy_score(y_test,pred_i))
plt.figure(figsize=(12, 6))
plt.plot(range(1, 5), acc, color='red', linestyle='dashed', marker='o',
markerfacecolor='blue', markersize=10)
plt.title('Accuracy vs p-Value')
plt.xlabel('p Value')
plt.ylabel('Accuracy')
```
