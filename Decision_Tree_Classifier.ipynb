```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
pima = pd.read_csv("co2.csv")
pima.head()
```

Removing Null Values and converting dataset to numeric type

```python
pima = pima.dropna()
pima = pima.reset_index()
for i in pima.columns:
    try:
        pima[[i]] = pima[[i]].astype(float).astype(int)
    except:
        pass
pima.head()
```

```python
feature_cols = ["Year", "Carbon_Dioxide","Carbon_Dioxide_Fit"]
x = pima[feature_cols]
y = pima.Carbon_Dioxide_Fit
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.3, random_state =1)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred))
print("Classification report - \n", classification_report(y_test,y_pred))
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(clf,out_file=dot_data,filled=True,rounded = True, special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('CO2.png')
Image(graph.create_png())
```
