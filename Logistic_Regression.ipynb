```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
pima = pd.read_csv("titanic.csv")
pima.head()
```

```python
pima = pima.dropna()
pima = pima.reset_index()
for i in pima.columns:
  try:
    pima[[i]] = pima[[i]].astype(float).astype(int)
  except:
    pass
```

Binomial Logistic Regression

```python
x = pima["Age"]
y = pima["Survived"]
```

```python
model = LogisticRegression()
```

```python
model.fit(x.values.reshape(-1, 1), y)
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict_proba(x.values.reshape(-1, 1))
```

```python
model.predict(x.values.reshape(-1, 1))
```

```python
model.score(x.values.reshape(-1, 1),y)
```

```python
confusion_matrix(y,model.predict(x.values.reshape(-1, 1)))
```

```python
print(classification_report(y,model.predict(x.values.reshape(-1, 1))))
```

Ordinal Logistic Regression

```python
ord_model = LogisticRegression(solver = 'liblinear', class_weight={0:0,1:1})
ord_model.fit(xb_train, yb_train)
```

```python
ord_model.score(xb_train,yb_train)
```

```python
yb_pr = ord_model.predict(xb_train)
```

Changing Parameters

Penalty

```python
bm = LogisticRegression(solver = 'liblinear', penalty= 'l2' , C=1.0)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

Dual

```python
bm = LogisticRegression(solver = 'liblinear', penalty= 'l2' , dual = True)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

max_iter

```python
bm = LogisticRegression(max_iter=1000)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

tol

```python
bm = LogisticRegression(solver = 'liblinear', tol= 1e-5, C=1.0)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

C

```python
bm = LogisticRegression(tol= 1e-5, C=10.0)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

fit_intecept

```python
bm = LogisticRegression(solver = 'liblinear',tol = 1e-4, C=20.0, fit_intercept = False)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

intercept_scaling

```python
bm = LogisticRegression(solver = 'liblinear',penalty='l2',intercept_scaling=14.8)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

class_weight

```python
bm = LogisticRegression(solver = 'liblinear', class_weight={0:3.8,1:0.2})
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

random_state

```python
bm = LogisticRegression(solver = 'sag',random_state=58)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

solver

```python
bm = LogisticRegression(solver = 'saga')
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

multi_class

```python
bm = LogisticRegression(multi_class='ovr')
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

verbose

```python
bm = LogisticRegression(solver = 'liblinear', verbose=5)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

l1_ratio

```python
bm = LogisticRegression(solver='saga', penalty='elasticnet', l1_ratio=0.66)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

n-jobs

```python
bm = LogisticRegression(n_jobs=10)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```

warm_start

```python
bm = LogisticRegression(solver='liblinear',warm_start= True)
bm.fit(x.values.reshape(-1,1),y)
bm.score(x.values.reshape(-1,1),y)
```
